```python
from matplotlib import pyplot as plt
```

```python
labels = ["Humans", "Smurfs", "Hobbits", "Ninjas"]
sizes = [220, 95, 80, 100]
```

```python
colors = ["yellow", "blue", "purple", "gray"]
explode = [.1, 0, 0, 0]
```

```python
plt.pie(sizes, labels=labels, colors=colors, explode=explode, shadow=True, startangle=-90, autopct="%1.1f%%")
```
