```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 2 * np.pi, 100)
sin_x = np.sin(x)
cos_x = np.cos(x)
```

```python
plt.plot(x, sin_x, label="Sin(x)", color="orange" )
plt.scatter(x, cos_x, label="Cos(x)", color="purple")
plt.xlabel("x")
plt.ylabel("y")
plt.title("A Touch of Trig")
plt.legend()
```
