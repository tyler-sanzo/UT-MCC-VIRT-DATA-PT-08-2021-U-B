```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 2 * np.pi, 100)
sin_x = np.sin(x)
cos_x = np.cos(x)
```

```python
plt.plot(x, sin_x, label = 'Sin(x)')
plt.scatter(x, cos_x, label = 'Cos(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.title('title')
plt.show()
```

```python
users = [13000, 26000, 52000, 3000, 9000]
labels = ['Java', 'C++', 'Python', 'Ruby', 'Clojure']
plt.bar(labels, users)
plt.xlabel('Language')
plt.ylabel('Users')
plt.ylim(0, 60000)
plt.show()
```
