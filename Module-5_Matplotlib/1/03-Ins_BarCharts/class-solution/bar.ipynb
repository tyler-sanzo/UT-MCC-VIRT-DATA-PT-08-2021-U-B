```python
import matplotlib.pyplot as plt
```

```python
users = [13000, 26000, 52000, 30000, 9000]
labels = ["Java", "C++", "Python", "Ruby", "Clojure"]
```

```python
plt.bar(labels, users)
plt.xlabel("Language")
plt.ylabel("Number of Users")
plt.title("Number of Users by Lang")
plt.ylim(0, max(users) * 1.25)
plt.show()
```
